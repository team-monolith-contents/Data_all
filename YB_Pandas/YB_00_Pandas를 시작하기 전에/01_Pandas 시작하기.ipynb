> Pandas라는 이름은 Panel Data에서 유래되었습니다.

> 각 데이터의 위치를 나타내기 위해서는 세로 방향의 순번(숫자)과 가로 방향의 순번(알파벳) 총 2가지를 알아야 합니다.  

###  **(1) Pandas 데이터의 2차원 구조**

> Pandas 데이터도 Panel과 같은 구조를 가집니다.  
> Pandas 데이터는 "행(가로)"과 "열(세로)" 두 가지 차원으로 데이터를 구분합니다.  

||1열|2열|
|---|---|---|
|**1행**| (1행,1열)| (1행,2열)|
|**2행**| (2행,1열)| (2행,2열)|

> 실생활의 많은 데이터가 2차원 구조의 형태를 지니고 있기 때문에 이를 위한 전용 패키지를 만들었습니다.  
> 이것이 바로 Pandas입니다.

### **(2) Pandas 데이터의 Key-Value 구조**

> Pandas의 데이터는 2차원 구조와 동시에 Key-Value의 구조를 가지고 있습니다.   
> 앞서 배웠던 딕셔너리(Dictionary) 데이터에서도 나오는 개념입니다.  
>
> Key-Value 구조의 데이터에서는 Key를 이용하여 Value를 빠르게 소환할 수 있어 커다란 데이터에 적합한 구조를 가지고 있습니다.  
>
> Pandas가 딕셔너리와 다른 점은 key와 value 구조를 지님과 동시에 List, Numpy array와 같이 순차적(번호순)으로 데이터를 불러올 수도 있다는 것입니다.  
>
> 한마디로 Key-Value 구조와 순차적 구조의 장점을 모두 가졌다! 고 볼 수 있죠.
