##  **1. Pandas로 데이터 출력**
### **(1) 데이터 불러오기**

> 우선 데이터가 있어야 그 내용을 확인할 수 있습니다.  
이번에도 seaborn 패키지에서 만들어진 데이터를 불러 오도록 하겠습니다. 
>
> `print(Sereis)` 또는 `print(DataFrame)` 으로 데이터를 출력해봅시다.

```python
# 예제 1: 제공되는 데이터 불러와서 출력해보기

import seaborn as sns
new_df = sns.load_dataset('planets')

print(new_df)
```

> `print` 함수는 데이터가 너무 큰 경우에 위와 같이 중간을 생략해서 출력합니다.  

> `print`를 이용해 출력한 것으로 알 수 있는 것은 무엇인가요?  
>-  method 열에는 Radial Velocity, Transit이란 값이 있다.  
>- number 열에는 1만 들어있다.  
>- mass 열에는 데이터 값이 없는 경우(NaN)가 있다.

### **(2) 데이터 원하는 만큼 출력하기**




> Pandas 데이터를 출력하는 또다른 방법에는 `head`, `tail`이 있습니다.  
> 이 방법들의 특징은 원하는 길이만큼을 출력할 수 있다는 점입니다. 
</br>  
> `Series.head(n)`, `DataFrame.head(n)`: 위에서부터 n개의 행을 출력  
> `Series.tail(n)`, `DataFrame.tail(n)`: 아래에서부터 n개의 행을 출력

```python
# 예제 2: 데이터의 앞순서부터 원하는 만큼 출력하기

print("데이터 앞에서부터 20개 출력:")
new_df.head(20)
```

> `head`를 이용해 출력한 것으로 더 알 수 있는 것은 무엇인가요?  
>- number 열에는 1보다 큰 값도 있다.   

```python
# 예제 3: 데이터의 뒷순서부터 원하는 만큼 출력하기

print("데이터 뒤에서부터 20개 출력:")
new_df.tail(20)
```

> `tail`을 이용해 출력한 것으로 더 알 수 있는 것은 무엇인가요?  
>- method 열에는 Imaging 이란 값도 있다.

## **2. Pandas로 데이터 정보 확인**

> 데이터의 일부분을 출력하는 것으로는 전체적인 데이터를 파악하기에 무리가 있습니다.  
> 데이터가 크다면 전체를 출력하여도 마찬가지입니다.  
>
> Pandas에는 데이터에 대한 전반적인 정보를 정리해주는 기능들이 있습니다.  

### (1) 데이터 기본 정보 확인하기
> `Series.info()`를 통해 다음을 알 수 있습니다.   
>- 총 행의 길이
>- Series의 이름
>- 데이터 갯수 (NaN 제외)
>- 데이터 타입
>- 데이터 용량

> `DataFrame.info()`를 통해 다음을 알 수 있습니다.   
>- 총 행의 길이
>- 각 열의 이름
>- 각 열에 존재하는 데이터 갯수 (NaN 제외)
>- 각 열의 데이터 타입
>- 데이터 용량

```python
# 예제 4: 데이터의 기본 정보 확인하기

new_df.info()
```

> `info`를 이용해 알아낸 정보를 정리해봅시다.
>- 총 행의 길이: 1035
>- method 열에는 1035개의 object타입 데이터가 있다.
>- number 열에는 1035개의 int타입 데이터가 있다.
>- orbital_period 열에는 992개의 float타입 데이터가 있다.
>- mass 열에는 513개의 float타입 데이터가 있다.
>- distance 열에는 808개의 float타입 데이터가 있다.
>- year 열에는 1035개의 int타입 데이터가 있다.
>- 데이터 용량: 약 48.6 KB

### (2) 데이터 통계 정보 확인하기
> `Sereis.describe()`, `DataFrame.describe()`를 통해 다음을 알 수 있습니다.   
>- count: 데이터 갯수 (NaN 제외)
>- mean: 평균값
>- std: 표준편차
>- min: 최솟값
>- 25%: 크기 순으로 정렬했을때 4분의 1번째에 해당하는 값
>- 50%: 크기 순으로 정렬했을때 중앙에 해당하는 값 (중앙값)
>- 75%: 크기 순으로 정렬했을때 4분의 3번째에 해당하는 값
>- max: 최댓값
>
> 위의 통계량은 숫자에 대한 것으로 데이터의 값이 어떻게 분포되어 있는지 알려줍니다.

```python
# 예제 5: 숫자 데이터의 통계 정보 확인하기

new_df.describe()
```

> `describe`를 통해 새롭게 알아낸 정보를 정리해봅시다.  
> - number 열에는 1부터 7까지의 값이 있다.
> - number 열의 최솟값, 중앙값이 1 이므로 절반 이상의 데이터가 1이다.
> - year 열에는 2007부터 2014까지의 값이 있다.

> 위의 결과에서 method 열의 데이터는 숫자가 아니기 때문에 통계량이 출력되지 않았습니다.   
> `PandasData.describe(include='O')` 를 이용하면 object 타입 데이터에 대한 다음 통계량을 알려줍니다.  
>- count: 데이터 갯수 (NaN 제외)
>- unique: 데이터의 고유값 종류 갯수
>- top: 가장 수가 많은 고유값
>- freq: top에 해당하는 고유값의 갯수

```python
# 예제 6: Object 타입 데이터의 통계 정보 확인하기

new_df.describe(include='O')
```

> 출력하기를 했을때는 method 열의 데이터를 총 3가지 밖에 확인하지 못했습니다.    
> `describe`를 이용하면 전체적으로 몇가지의 고유값이 있는지 쉽게 알 수 있습니다.  
>
> 새로 알아낸 정보를 정리해봅시다.
>- method 열에는 총 10가지 종류의 값이 있다.
>- Radial Velocity가 가장 많고 1035개 데이터 중 553개를 차지한다.
