```python
import pandas as pd  # ÌåêÎã§Ïä§Î•º pdÎ°ú Ï∂ïÏïΩÌïòÏó¨ Î∂àÎü¨ÏòµÎãàÎã§.
import matplotlib.pyplot as plt  # PyplotÏùÑ pltÎ°ú Ï∂ïÏïΩÌïòÏó¨ Î∂àÎü¨ÏòµÎãàÎã§.
```

# **[11] Í∑∏ÎûòÌîÑ Ïó¨Îü¨Í∞ú ÎÇòÏó¥ÌïòÍ∏∞**

[Ïò§Îäò Î∞∞Ïö∏ Í≤É]
1. Ïó¨Îü¨ Í∑∏ÎûòÌîÑÎ•º ÏÑ∏Î°úÎ∞©Ìñ•ÏúºÎ°ú ÎÇòÏó¥Ìï† Ïàò ÏûàÎã§.

2. Ïó¨Îü¨ Í∑∏ÎûòÌîÑÎ•º 2Ï∞®ÏõêÏúºÎ°ú ÎÇòÏó¥Ìï† Ïàò ÏûàÎã§.  


##  **1. Í∑∏ÎûòÌîÑ Ïó¨Îü¨Í∞ú ÎÇòÏó¥ÌïòÍ∏∞**
  
> ÌïòÎÇòÏùò Í∑∏Î¶º(figure)Ïóê Ïó¨Îü¨Í∞úÏùò Ï¢åÌëúÏ∂ï(axes)ÏùÑ ÎÑ£Ïñ¥ Í∑∏ÎûòÌîÑ Ïó¨Îü¨Í∞úÎ•º Ìï®Íªò Í∑∏Î¶¥ Ïàò ÏûàÏäµÎãàÎã§.  
>
> `subplots` Ìï®ÏàòÎäî figureÏôÄ axes Ïó¨Îü¨Í∞úÎ•º Î∞òÌôòÌï¥Ï§çÎãàÎã§.  
> Î∞òÌôòÎêú figureÏôÄ axesÎ•º Ïù¥Ïö©ÌïòÏó¨ Í∑∏ÎûòÌîÑÎ•º Íæ∏Î©∞Ï§Ñ Ïàò ÏûàÏäµÎãàÎã§.  
>
> figureÎ•º Ïù¥Ïö©ÌïòÏó¨ Ï†ÑÏ≤¥ Í∑∏Î¶ºÏùò Ï†úÎ™©, Î∞îÌÉïÏÉâ, ÌÖåÎëêÎ¶¨Î•º ÏÑ§Ï†ïÌï† Ïàò ÏûàÏäµÎãàÎã§.  
> axesÎ•º Ïù¥Ïö©ÌïòÏó¨ Í∞Å Í∑∏ÎûòÌîÑÏùò Ï†úÎ™©, ÎààÍ∏à, ÏÉâÏÉÅ, Í∑∏ÎûòÌîÑ, Î≤îÎ°Ä Îì±ÏùÑ ÏÑ§Ï†ï Ìï† Ïàò ÏûàÏäµÎãàÎã§.  
>
> ÏÑ§Ï†ïÏùÑ ÏúÑÌïú Ìï®Ïàò Ïù¥Î¶ÑÏóê Ï£ºÏùòÌïòÎ©∞ Í∑∏ÎûòÌîÑÎ•º Ïó¨Îü¨Í∞ú ÎÇòÏó¥ÌïòÎäî Î∞©Î≤ïÏùÑ Î∞∞ÏõåÎ¥ÖÏãúÎã§.  


##  **2. Í∑∏ÎûòÌîÑ ÏÑ∏Î°úÎ∞©Ìñ•ÏúºÎ°ú ÎÇòÏó¥ÌïòÍ∏∞**



### **(1) subplotsÌï®ÏàòÍ∞Ä Î∞òÌôòÌïòÎäî figureÏôÄ axes Î¶¨Ïä§Ìä∏**

> Îã§ÏùåÍ≥º Í∞ôÏù¥ `subplots` Ìï®ÏàòÏóê figureÏóê Ï∂îÍ∞ÄÌï† axesÏóê Í∞úÏàòÎ•º ÎÑ£Ïñ¥ ÏÑ∏Î°úÎ∞©Ìñ•ÏúºÎ°ú ÎÇòÏó¥Îêú axesÏôÄ figureÎ•º ÏñªÏùÑ Ïàò ÏûàÏäµÎãàÎã§.  
>
> `figure, axes = plt.subplots(N, figsize=(w, h))`   
> : NÍ∞úÏùò axesÍ∞Ä ÏÑ∏Î°úÎ∞©Ìñ•ÏúºÎ°ú ÎÇòÏó¥ÎêòÏñ¥ ÏûàÍ≥† ÎÑàÎπÑÍ∞Ä w Ïù∏Ïπò, ÎÜíÏù¥Í∞Ä h Ïù∏ÏπòÏù∏ figureÏôÄ axesÎ¶¨Ïä§Ìä∏Î•º Î∞òÌôò  
>  
> ÏúÑÏóêÏÑú ÏñªÏùÄ axesÎäî Ïó¨Îü¨Í∞úÏùò Í∑∏ÎûòÌîÑ(axes)Î•º Ìè¨Ìï®ÌïòÎäî Î¶¨Ïä§Ìä∏ÏûÖÎãàÎã§.  
> Îî∞ÎùºÏÑú Í∞Å Í∑∏ÎûòÌîÑ(axes)Î•º Íæ∏Î∞Ä ÎïåÏóêÎäî `axes[i]` ÏôÄ Í∞ôÏù¥ ÏÑ∏Î°úÎ∞©Ìñ• ÏàúÏÑúÎ•º Ïù¥Ïö©Ìï¥ Ïù∏Îç±Ïã±ÌïòÏó¨ Î∂àÎü¨ÏôÄ ÏÇ¨Ïö©Ìï©ÎãàÎã§.  
>
> Îã§Ïùå ÏòàÏ†úÎ•º Ïã§ÌñâÌïòÏó¨ Í∞Å Í∑∏ÎûòÌîÑ(axes)Ïùò Î≤àÌò∏Î•º ÌôïÏù∏ÌïòÏó¨ Î¥ÖÏãúÎã§. 

```python
# ÏòàÏ†ú 1: Ïó¨Îü¨ axesÍ∞Ä ÏÑ∏Î°úÎ∞©Ìñ•ÏúºÎ°ú ÎÇòÏó¥Îêú figure ÌôïÏù∏ÌïòÍ∏∞

fig, axes = plt.subplots(2, figsize=(10, 6))  # 2Í∞úÏùò axesÍ∞Ä ÏÑ∏Î°úÎ∞©Ìñ•ÏúºÎ°ú ÎÇòÏó¥ÎêòÏñ¥ ÏûàÍ≥†, ÎÑàÎπÑ 10Ïù∏Ïπò ÎÜíÏù¥ 6Ïù∏ÏπòÏù∏ figureÏôÄ axesÎ¶¨Ïä§Ìä∏Î•º Î∞òÌôò

fig.suptitle("Í∑∏ÎûòÌîÑ Ïó¨Îü¨Í∞ú ÎÇòÏó¥ÌïòÍ∏∞", fontsize=20)
for i in range(2):
    axes[i].set_title("axes["+str(i)+"]")
plt.show()
```

### **(2) subplotsÌï®ÏàòÏôÄ figure Íæ∏ÎØ∏Í∏∞**

> Îã§ÏùåÍ≥º Í∞ôÏù¥ `subplots` Ìï®ÏàòÏóê figureÏóê Ï∂îÍ∞ÄÌï† axesÏóê Í∞úÏàòÎ•º ÎÑ£Ïñ¥ ÏÑ∏Î°úÎ∞©Ìñ•ÏúºÎ°ú ÎÇòÏó¥Îêú axesÏôÄ figureÎ•º ÏñªÏùÑ Ïàò ÏûàÏäµÎãàÎã§.  
> Ïù¥Îïå, figureÏùò ÌÅ¨Í∏∞, Î∞îÌÉïÏÉâ, Î™®ÏÑúÎ¶¨ ÏÉâ, Î™®ÏÑúÎ¶¨ ÍµµÍ∏∞Î•º ÏÑ§Ï†ïÌï† Ïàò ÏûàÏäµÎãàÎã§.  
>
> * **figure size** : `figsize=(w, h)` Î•º Ïù¥Ïö©Ìï¥ ÎÑàÎπÑ wÏù∏Ïπò, ÎÜíÏù¥ hÏù∏ÏπòÎ°ú ÏßÄÏ†ï  
> * **face color** : `facecolor` Ïù∏ÏàòÏóê Ïù¥Î¶ÑÏùÑ ÎÑ£Ïñ¥ Î∞îÌÉïÏÉâ ÏßÄÏ†ï  
> * **edge color** : `edgecolor` Ïù∏ÏàòÏóê Ïù¥Î¶ÑÏùÑ ÎÑ£Ïñ¥ Î™®ÏÑúÎ¶¨ ÏÉâ ÏßÄÏ†ï  
> * **edge line width** : `linewidth` Ïù∏ÏàòÏóê Ïà´ÏûêÎ•º ÎÑ£Ïñ¥ Î™®ÏÑúÎ¶¨ ÍµµÍ∏∞ ÏßÄÏ†ï
>
> ÎòêÌïú, Ïó¨Îü¨ axesÎì§Ïùò Ï∂ïÏùÑ Í≥µÏú†Ìï† Í≤ÉÏù∏ÏßÄ ÏÑ§Ï†ïÌï† Ïàò ÏûàÏäµÎãàÎã§.  
>
> * **share x** : `sharex` Ïù∏ÏàòÏóê True/FalseÎ•º ÎÑ£Ïñ¥ xÏ∂ïÏùÑ Í≥µÏú†Ìï† Í≤ÉÏù∏ÏßÄ ÏßÄÏ†ï  
> * **share y** : `sharey` Ïù∏ÏàòÏóê True/FalseÎ•º ÎÑ£Ïñ¥ yÏ∂ïÏùÑ Í≥µÏú†Ìï† Í≤ÉÏù∏ÏßÄ ÏßÄÏ†ï  

> figureÏùò Ï†úÎ™©ÏùÄ Îã§ÏùåÍ≥º Í∞ôÏù¥ `suptitle` Ìï®ÏàòÎ•º Ïù¥Ïö©ÌïòÏó¨ ÏÑ§Ï†ïÌï† Ïàò ÏûàÏäµÎãàÎã§.  
>
> `figure.suptitle("Ï†ÑÏ≤¥ Ï†úÎ™©")` : "Ï†ÑÏ≤¥ Ï†úÎ™©"Ïù¥ÎùºÎäî Ï†úÎ™©ÏùÑ Ï∂îÍ∞Ä
> * **color** : `color` Ïù∏ÏàòÏóê Ïù¥Î¶ÑÏùÑ ÎÑ£Ïñ¥ Í∏ÄÏûêÏÉâ ÏßÄÏ†ï
> * **fontsize** : `fontsize` Ïù∏ÏàòÏóê Ïà´ÏûêÎ•º ÎÑ£Ïñ¥ Í∏ÄÏûêÌÅ¨Í∏∞ ÏßÄÏ†ï
> * **fontweight** : `fontweight` Ïù∏ÏàòÏóê Ïà´Ïûê ÌòπÏùÄ normal/boldÎ•º ÎÑ£Ïñ¥ Í∏ÄÏûêÍµµÍ∏∞ ÏßÄÏ†ï

```python
# ÏòàÏ†ú 2: Ïó¨Îü¨ axesÍ∞Ä ÏÑ∏Î°úÎ∞©Ìñ•ÏúºÎ°ú ÎÇòÏó¥Îêú figure Íæ∏ÎØ∏Í∏∞

# 2Í∞úÏùò axesÍ∞Ä ÏÑ∏Î°úÎ∞©Ìñ•ÏúºÎ°ú ÎÇòÏó¥ÎêòÏñ¥ ÏûàÍ≥†, ÎÑàÎπÑ 10Ïù∏Ïπò ÎÜíÏù¥ 6Ïù∏ÏπòÏù∏ figureÏôÄ axesÎ¶¨Ïä§Ìä∏Î•º Î∞òÌôò
# figureÏùò Î∞îÌÉïÏÉâÏùÄ ÎÖ∏Îûë, Î™®ÏÑúÎ¶¨ ÏÉâÏùÄ Í≤ÄÏ†ï, Î™®ÏÑúÎ¶¨ ÍµµÍ∏∞Îäî 2Î°ú ÏÑ§Ï†ï
# axesÎì§Ïùò xÏ∂ïÏùÑ Í≥µÏú†
fig, axes = plt.subplots(2, figsize=(10, 6), facecolor='yellow', edgecolor='k', linewidth=2, sharey=True)

# Í∏ÄÏûêÌÅ¨Í∏∞ 20, bold ÍµµÍ∏∞Ïùò Ï†ÑÏ≤¥ Ï†úÎ™© Ï∂îÍ∞Ä
fig.suptitle("Í∑∏ÎûòÌîÑ Ïó¨Îü¨Í∞ú ÎÇòÏó¥ÌïòÍ∏∞", fontsize=20, fontweight='bold')

# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
for i in range(2):
    axes[i].set_title("axes["+str(i)+"]")

# axes[0]Ïùò x, yÏ∂ï Î≤îÏúÑ ÏÑ§Ï†ï
axes[0].set_xlim(-1, 1)
axes[0].set_ylim(-1, 1)

plt.show()  # figureÎ•º Ï∂úÎ†•
```

> ÏúÑÏùò Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶¥ Îïå axesÎì§Ïùò xÏ∂ïÏùÄ Í≥µÏú†ÌïòÏßÄ ÏïäÏïòÍ≥†, yÏ∂ïÏùÄ Í≥µÏú†ÌïòÏòÄÏäµÎãàÎã§.  
> axes[0]Ïùò Ï∂ï Î≤îÏúÑÎ•º Í∏∞Î≥∏ ÏÑ§Ï†ïÍ≥º Îã§Î•¥Í≤å ÏÑ§Ï†ïÌïòÏòÄÎçîÎãà Îëê Í∑∏ÎûòÌîÑÏùò xÏ∂ïÏùÄ Îã§Î•¥ÏßÄÎßå yÏ∂ïÏùÄ Î™®Îëê ÏÑ§Ï†ïÏù¥ Î∞îÎÄåÏóàÏùåÏùÑ ÌôïÏù∏ÌïòÏó¨ Î¥ÖÏãúÎã§. 

### **(3) Í∑∏ÎûòÌîÑ(axes) Íæ∏ÎØ∏Í∏∞**

> axesÎ¶¨Ïä§Ìä∏ ÏïàÏùò axes[i]Î•º Ïù¥Ïö©ÌïòÏó¨ Í∞Å Í∑∏ÎûòÌîÑÎ•º Íæ∏Î∞Ä Ïàò ÏûàÏäµÎãàÎã§.  
>
> * **Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞** : Í∫æÏùÄÏÑ†Í∑∏ÎûòÌîÑ, ÏÇ∞Ï†êÎèÑ Îì± Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶¨Îäî Ìï®ÏàòÎäî Ïù¥Ï†ÑÏóê Î∞∞Ïö¥Í≤ÉÍ≥º ÎèôÏùºÌïòÎÇò ÏÇ¨Ïö©ÌïòÎäî Î∞©ÏãùÏù¥ Îã§Î¶ÖÎãàÎã§.  
>
>> `axes[i].plot()` : (i-1)Î≤àÏß∏ Í∑∏ÎûòÌîÑÏóê Í∫æÏùÄÏÑ† Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶º
>
> * **Í∑∏ÎûòÌîÑ Íæ∏ÎØ∏Í∏∞** : Ï†úÎ™©, ÎààÍ∏à, Í≤©Ïûê Îì±ÏùÑ ÏÑ§Ï†ïÌïòÎäî Ìï®ÏàòÎäî Ïù¥Î¶ÑÏù¥ ÏïΩÍ∞Ñ Îã§Î•¥Îãà Ï£ºÏùòÌï¥Ïïº Ìï©ÎãàÎã§.  
>
>> `axes[i].set_title()` : (i-1)Î≤àÏß∏ Í∑∏ÎûòÌîÑÏùò Ï†úÎ™©ÏùÑ ÏÑ§Ï†ï (title Ìï®ÏàòÏôÄ Ïù∏ÏàòÍ∞Ä Í∞ôÏùå)  
>>
>> `axes[i].set_xticks()` : (i-1)Î≤àÏß∏ Í∑∏ÎûòÌîÑÏùò xÏ∂ï ÎààÍ∏àÏùÑ ÏÑ§Ï†ï (xticks Ìï®ÏàòÏôÄ Ïù∏ÏàòÍ∞Ä Í∞ôÏùå)  
>> `axes[i].set_yticks()` : (i-1)Î≤àÏß∏ Í∑∏ÎûòÌîÑÏùò yÏ∂ï ÎààÍ∏àÏùÑ ÏÑ§Ï†ï (yticks Ìï®ÏàòÏôÄ Ïù∏ÏàòÍ∞Ä Í∞ôÏùå)
>>
>> `axes[i].set_xlabel()` : (i-1)Î≤àÏß∏ Í∑∏ÎûòÌîÑÏùò xÏ∂ï Ïù¥Î¶ÑÏùÑ ÏÑ§Ï†ï (xlabel Ìï®ÏàòÏôÄ Ïù∏ÏàòÍ∞Ä Í∞ôÏùå)  
>> `axes[i].set_ylabel()` : (i-1)Î≤àÏß∏ Í∑∏ÎûòÌîÑÏùò yÏ∂ï Ïù¥Î¶ÑÏùÑ ÏÑ§Ï†ï (ylabel Ìï®ÏàòÏôÄ Ïù∏ÏàòÍ∞Ä Í∞ôÏùå)
>>
>> `axes[i].grid()` : (i-1)Î≤àÏß∏ Í∑∏ÎûòÌîÑÏùò Í≤©Ïûê ÏÑ§Ï†ï (grid Ìï®ÏàòÏôÄ Ïù∏ÏàòÍ∞Ä Í∞ôÏùå)
>>
>> `axes[i].set_facecolor()` : (i-1)Î≤àÏß∏ Í∑∏ÎûòÌîÑÏùò Î∞îÌÉïÏÉâ ÏÑ§Ï†ï

```python
# ÏòàÏ†ú 3: Ïó¨Îü¨ axes Íæ∏ÎØ∏Í∏∞

## (1) figure ÏÉùÏÑ±
# 2Í∞úÏùò axesÍ∞Ä ÏÑ∏Î°úÎ∞©Ìñ•ÏúºÎ°ú ÎÇòÏó¥ÎêòÏñ¥ ÏûàÍ≥†, ÎÑàÎπÑ 10Ïù∏Ïπò ÎÜíÏù¥ 8Ïù∏Ïπò ÎÖ∏ÎûÄ Î∞îÌÉïÏÉâÏùò figureÎ•º Î∞òÌôò
# ÎòêÌïú xÏ∂ïÍ≥º yÏ∂ïÏù¥ Í∞ôÏùÄ axesÎ¶¨Ïä§Ìä∏Î•º Î∞òÌôò
fig, axes = plt.subplots(2, figsize=(10, 8), facecolor='yellow', sharex=True, sharey=True)

## (2) figure Ï†úÎ™© ÏÑ§Ï†ï
fig.suptitle('Ïó¨Îü¨ Í∑∏ÎûòÌîÑ ÏÑ∏Î°úÎ∞©Ìñ•ÏúºÎ°ú ÎÇòÏó¥ÌïòÍ∏∞', fontsize=20, fontweight='bold')  # Í∏ÄÏûêÌÅ¨Í∏∞ 20, bold ÍµµÍ∏∞Ïùò Ï†ÑÏ≤¥ Ï†úÎ™©ÏùÑ Ï∂îÍ∞Ä


## (3) axes[0] Í∑∏ÎûòÌîÑ ÏÑ§Ï†ï
# Îç∞Ïù¥ÌÑ∞
import numpy as np
X1 = np.arange(-1, 16, 0.2)  # -1 Î∂ÄÌÑ∞ 0.2Í∞ÑÍ≤©ÏúºÎ°ú 16 ÎØ∏ÎßåÍπåÏßÄÏùò ÏàòÎ•º arrayÏóê Ï†ÄÏû•
Y1 = np.sin(X1)  # X1 ÏïàÏùò Ïàò xÏóê ÎåÄÌï¥ÏÑú sin(x) Í∞íÏùÑ arrayÏóê Ï†ÄÏû•
# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
axes[0].plot(X1, Y1)  # axes[0]Ïóê Í∫æÏùÄÏÑ† Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶º
# Í∑∏ÎûòÌîÑ Íæ∏ÎØ∏Í∏∞
axes[0].set_title('y = sin(x)', fontsize=20)  # Ï†úÎ™© ÏÑ§Ï†ï
axes[0].set_xticks(np.pi*0.5*np.arange(10), ['0', '0.5pi', 'pi', '1.5pi', '2pi', '2.5pi', '3pi', '3.5pi', '4pi', '4.5pi'])  # ÎààÍ∏à ÏÑ§Ï†ï (axes[1]Í≥º Í≥µÏú†)
axes[0].set_ylabel("y", rotation='horizontal')  # yÏ∂ï Ïù¥Î¶Ñ ÏÑ§Ï†ï
axes[0].grid(axis='both', color='purple', linestyle=':')  # Í≤©Ïûê ÏÑ§Ï†ï
axes[0].set_facecolor('lime')  # Î∞îÌÉïÏÉâ ÏÑ§Ï†ï


## (4) axes[1] Í∑∏ÎûòÌîÑ ÏÑ§Ï†ï
# Îç∞Ïù¥ÌÑ∞
X2 = np.arange(-1, 16, 0.2)  # -1 Î∂ÄÌÑ∞ 0.2Í∞ÑÍ≤©ÏúºÎ°ú 16 ÎØ∏ÎßåÍπåÏßÄÏùò ÏàòÎ•º arrayÏóê Ï†ÄÏû•
Y2 = np.cos(X2)  # X2 ÏïàÏùò Ïàò xÏóê ÎåÄÌï¥ÏÑú cos(x) Í∞íÏùÑ arrayÏóê Ï†ÄÏû•
# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
axes[1].plot(X2, Y2)  # axes[1]Ïóê Í∫æÏùÄÏÑ† Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶º
# Í∑∏ÎûòÌîÑ Íæ∏ÎØ∏Í∏∞
axes[1].set_title('y = cos(x)', fontsize=20)  # Ï†úÎ™© ÏÑ§Ï†ï
axes[1].set_ylabel("y", rotation='horizontal')  # yÏ∂ï Ïù¥Î¶Ñ ÏÑ§Ï†ï
axes[1].set_xlabel("x")  # xÏ∂ï Ïù¥Î¶Ñ ÏÑ§Ï†ï
axes[1].set_xlim(-1, 15)  # xÏ∂ï Î≤îÏúÑ ÏÑ§Ï†ï (axes[0]ÏôÄ Í≥µÏú†)
axes[1].grid(axis='both', color='purple', linestyle=':')  # Í≤©Ïûê ÏÑ§Ï†ï

plt.show()  # figureÎ•º Ï∂úÎ†•
```

> Îî∞ÎùºÏÑú Í∞Å Í∑∏ÎûòÌîÑ(axes)Î•º Íæ∏Î∞Ä ÎïåÏóêÎäî `axes[j]` ÏôÄ Í∞ôÏù¥ Í∞ÄÎ°úÎ∞©Ìñ• ÏàúÏÑúÎ•º Ïù¥Ïö©Ìï¥ Ïù∏Îç±Ïã±ÌïòÏó¨ Î∂àÎü¨ÏôÄ ÏÇ¨Ïö©Ìï©ÎãàÎã§.  
>
> Îã§Ïùå ÏòàÏ†úÎ•º Ïã§ÌñâÌïòÏó¨ Í∞Å Í∑∏ÎûòÌîÑ(axes)Ïùò Î≤àÌò∏Î•º ÌôïÏù∏ÌïòÏó¨ Î¥ÖÏãúÎã§. 

```python
# ÏòàÏ†ú 4: Í∑∏ÎûòÌîÑ Ïó¨Îü¨Í∞ú ÎÇòÏó¥Îêú figure ÌôïÏù∏ÌïòÍ∏∞

fig, axes = plt.subplots(1, 3, figsize=(10, 6))  # 3Í∞úÏùò axesÍ∞Ä Í∞ÄÎ°úÎ∞©Ìñ•ÏúºÎ°ú ÎÇòÏó¥ÎêòÏñ¥ ÏûàÍ≥†, ÎÑàÎπÑ 10Ïù∏Ïπò ÎÜíÏù¥ 6Ïù∏ÏπòÏù∏ figureÏôÄ axesÎ¶¨Ïä§Ìä∏Î•º Î∞òÌôò

fig.suptitle("Í∑∏ÎûòÌîÑ Ïó¨Îü¨Í∞ú ÎÇòÏó¥ÌïòÍ∏∞", fontsize=20)
for j in range(3):
    axes[j].set_title("axes["+str(j)+"]")
plt.show()
```

### **(2) figureÏôÄ Ïó¨Îü¨Í∞ú axes Íæ∏ÎØ∏Í∏∞**

> ÏÑ∏Î°úÎ∞©Ìñ•, Í∞ÄÎ°úÎ∞©Ìñ•ÏúºÎ°ú ÎÇòÏó¥Îêú Í∑∏ÎûòÌîÑÎ•º Íæ∏ÎØ∏Îäî Î∞©Î≤ïÏùÄ Í∞ôÏäµÎãàÎã§.  
>
> ÏïÑÎûò ÏòàÏ†úÎ•º ÌÜµÌï¥ Í∞ÄÎ°úÎ∞©Ìñ•ÏúºÎ°ú ÎÇòÏó¥Îêú ÏõêÍ∑∏ÎûòÌîÑÎ•º Í∑∏Î†§Î¥ÖÏãúÎã§. 

```python
# ÏòàÏ†ú 5: figureÏôÄ Ïó¨Îü¨ axes Íæ∏ÎØ∏Í∏∞

## (1) figure ÏÉùÏÑ±
# 3Í∞úÏùò axesÍ∞Ä Í∞ÄÎ°úÎ∞©Ìñ•ÏúºÎ°ú ÎÇòÏó¥ÎêòÏñ¥ ÏûàÍ≥†, ÎÑàÎπÑ 16Ïù∏Ïπò ÎÜíÏù¥ 6Ïù∏Ïπò ÍµµÍ∏∞ 3, Í≤ÄÏ†ïÏÉâ Î™®ÏÑúÎ¶¨Ïùò figureÎ•º Î∞òÌôò
fig, axes = plt.subplots(1, 3, figsize=(16,6), edgecolor='k', linewidth=3)

## (2) figure Ï†úÎ™© ÏÑ§Ï†ï
fig.suptitle('Ïó¨Îü¨ Í∑∏ÎûòÌîÑ Í∞ÄÎ°úÎ∞©Ìñ•ÏúºÎ°ú ÎÇòÏó¥ÌïòÍ∏∞', fontsize=20, fontweight='bold')

# ÏõêÍ∑∏ÎûòÌîÑ ÌäπÏßï ÏÑ§Ï†ï
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 1}
# ÏõêÏùò Ï§ëÏã¨ÏúºÎ°ú Î∂ÄÌÑ∞ Î∞òÏßÄÎ¶ÑÏùò 30%Ïóê Ìï¥ÎãπÌïòÎäî Î∂ÄÎ∂ÑÏùÑ ÏûòÎùºÎÉÖÎãàÎã§. 
# Î∂ÄÏ±ÑÍº¥Ïóê ÎëêÍªòÍ∞Ä 1pixelÏù¥Í≥† ÏÉâÏÉÅÏùÄ ÌïòÏñÄ Î™®ÏÑúÎ¶¨ ÏÑ†ÏùÑ Í∑∏Î¶ΩÎãàÎã§.

## (3) axes[0] Í∑∏ÎûòÌîÑ ÏÑ§Ï†ï
# Îç∞Ïù¥ÌÑ∞
AÎ∞òÌõÑÎ≥¥ = ['1Î≤à', '2Î≤à', '3Î≤à', '4Î≤à', '5Î≤à', '6Î≤à', '7Î≤à']
AÎ∞òÎìùÌëú = [4, 2, 1, 9, 8, 0, 1]
# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
axes[0].pie(AÎ∞òÎìùÌëú, labels=AÎ∞òÌõÑÎ≥¥, autopct='%.1f', wedgeprops=wedgeprops)  # axes[0]Ïóê ÏõêÍ∑∏ÎûòÌîÑÎ•º Í∑∏Î¶º
# Ï†úÎ™© ÏÑ§Ï†ï
axes[0].set_title("AÎ∞ò", fontsize=20) 

## (4) axes[1] Í∑∏ÎûòÌîÑ ÏÑ§Ï†ï
# Îç∞Ïù¥ÌÑ∞
BÎ∞òÌõÑÎ≥¥ = ['1Î≤à', '2Î≤à', '3Î≤à', '4Î≤à', '5Î≤à']
BÎ∞òÎìùÌëú = [14, 4, 2, 2, 3]
# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
axes[1].pie(BÎ∞òÎìùÌëú, labels=BÎ∞òÌõÑÎ≥¥, autopct='%.1f', wedgeprops=wedgeprops)  # axes[1]Ïóê ÏõêÍ∑∏ÎûòÌîÑÎ•º Í∑∏Î¶º
# Ï†úÎ™© ÏÑ§Ï†ï
axes[1].set_title("BÎ∞ò", fontsize=20) 

## (5) axes[2] Í∑∏ÎûòÌîÑ ÏÑ§Ï†ï
# Îç∞Ïù¥ÌÑ∞
CÎ∞òÌõÑÎ≥¥ = ['1Î≤à', '2Î≤à', '3Î≤à', '4Î≤à', '5Î≤à', '6Î≤à']
CÎ∞òÎìùÌëú = [1, 6, 0, 9, 7, 2]
# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
axes[2].pie(CÎ∞òÎìùÌëú, labels=CÎ∞òÌõÑÎ≥¥, autopct='%.1f', wedgeprops=wedgeprops)  # axes[2]Ïóê ÏõêÍ∑∏ÎûòÌîÑÎ•º Í∑∏Î¶º
# Ï†úÎ™© ÏÑ§Ï†ï
axes[2].set_title("CÎ∞ò", fontsize=20) 

plt.show()  # figureÎ•º Ï∂úÎ†•
```

##  **3. Í∑∏ÎûòÌîÑ 2Ï∞®ÏõêÏúºÎ°ú ÎÇòÏó¥ÌïòÍ∏∞**


### **(1) subplotsÌï®ÏàòÍ∞Ä Î∞òÌôòÌïòÎäî figureÏôÄ axes Î¶¨Ïä§Ìä∏**

> Îã§ÏùåÍ≥º Í∞ôÏù¥ `subplots` Ìï®ÏàòÏóê figureÏóê Ï∂îÍ∞ÄÌï† axesÏóê Í∞úÏàòÎ•º ÎÑ£Ïñ¥ 2Ï∞®ÏõêÏúºÎ°ú ÎÇòÏó¥Îêú axesÏôÄ figureÎ•º ÏñªÏùÑ Ïàò ÏûàÏäµÎãàÎã§.  
>
> `figure, axes = plt.subplots(N, M, figsize=(w, h))`   
> : ÏÑ∏Î°úÎ∞©Ìñ•ÏúºÎ°ú 2Í∞ú, Í∞ÄÎ°úÎ∞©Ìñ•ÏúºÎ°ú 3Í∞ú Ï¥ù 6Í∞úÏùò axesÍ∞Ä ÎÇòÏó¥Îêú ÎÑàÎπÑÍ∞Ä w Ïù∏Ïπò, ÎÜíÏù¥Í∞Ä h Ïù∏ÏπòÏù∏ figureÏôÄ axesÎ¶¨Ïä§Ìä∏Î•º Î∞òÌôò  
>  
> ÏúÑÏóêÏÑú ÏñªÏùÄ axesÎäî Ïó¨Îü¨Í∞úÏùò Í∑∏ÎûòÌîÑ(axes)Î•º Ìè¨Ìï®ÌïòÎäî Î¶¨Ïä§Ìä∏ÏûÖÎãàÎã§.  
> Îî∞ÎùºÏÑú Í∞Å Í∑∏ÎûòÌîÑ(axes)Î•º Íæ∏Î∞Ä ÎïåÏóêÎäî `axes[i, j]` ÏôÄ Í∞ôÏù¥ ÏàúÏÑúÎ•º 2Ï∞®Ïõê Ï¢åÌëúÎ•º Ïù¥Ïö©Ìï¥ Ïù∏Îç±Ïã±ÌïòÏó¨ Î∂àÎü¨ÏôÄ ÏÇ¨Ïö©Ìï©ÎãàÎã§.  
>
> Îã§Ïùå ÏòàÏ†úÎ•º Ïã§ÌñâÌïòÏó¨ Í∞Å Í∑∏ÎûòÌîÑ(axes)Ïùò Î≤àÌò∏Î•º ÌôïÏù∏ÌïòÏó¨ Î¥ÖÏãúÎã§. 

```python
# ÏòàÏ†ú 6: Ïó¨Îü¨ axesÍ∞Ä 2Ï∞®ÏõêÏúºÎ°ú ÎÇòÏó¥Îêú figure ÌôïÏù∏ÌïòÍ∏∞

fig, axes = plt.subplots(2, 3, figsize=(10, 6))  # 2Í∞úÏùò axesÍ∞Ä ÏÑ∏Î°úÎ∞©Ìñ•ÏúºÎ°ú ÎÇòÏó¥ÎêòÏñ¥ ÏûàÍ≥†, ÎÑàÎπÑ 10Ïù∏Ïπò ÎÜíÏù¥ 6Ïù∏ÏπòÏù∏ figureÏôÄ axesÎ¶¨Ïä§Ìä∏Î•º Î∞òÌôò

fig.suptitle("Í∑∏ÎûòÌîÑ Ïó¨Îü¨Í∞ú ÎÇòÏó¥ÌïòÍ∏∞", fontsize=20)

for i in range(2):
    for j in range(3):
        axes[i, j].set_title("axes["+str(i)+", "+str(j)+"]")
plt.show()
```

### **(2) figureÏôÄ Ïó¨Îü¨Í∞ú axes Íæ∏ÎØ∏Í∏∞**

> 2Ï∞®ÏõêÏúºÎ°ú ÎÇòÏó¥Îêú Í∑∏ÎûòÌîÑÎ•º Íæ∏ÎØ∏Îäî Î∞©Î≤ïÎèÑ Îã§Î•¥ÏßÄ ÏïäÏäµÎãàÎã§.  
>
> Ïù¥Î≤àÏóêÎäî forÎ¨∏ÏùÑ Ïù¥Ïö©ÌïòÏó¨ Í∞Å Í∑∏ÎûòÌîÑÏóê Ï†ëÍ∑ºÌïòÍ≥† ÌûàÏä§ÌÜ†Í∑∏Îû®ÏùÑ Í∑∏Î†§Î≥¥Í≤†ÏäµÎãàÎã§. 

```python
# ÏòàÏ†ú 7: figureÏôÄ Ïó¨Îü¨ axes Íæ∏ÎØ∏Í∏∞, forÎ¨∏ Ïù¥Ïö©ÌïòÍ∏∞

## (1) figure ÏÉùÏÑ±
# axesÍ∞Ä Í∞ÄÎ°úÎ∞©Ìñ• 2Ï§Ñ ÏÑ∏Î°úÎ∞©Ìñ• 3Ï§ÑÎ°ú Ï¥ù 6Í∞ú ÎÇòÏó¥ÎêòÏñ¥ ÏûàÍ≥†, ÎÑàÎπÑ 16Ïù∏Ïπò ÎÜíÏù¥ 10Ïù∏ÏπòÏùò figureÎ•º Î∞òÌôò
# ÎòêÌïú xÏ∂ïÍ≥º yÏ∂ïÏù¥ Í∞ôÏùÄ axesÎ¶¨Ïä§Ìä∏Î•º Î∞òÌôò
fig, axes = plt.subplots(2, 3, figsize=(16, 10), sharex=True, sharey=True)

## (2) figure Ï†úÎ™© ÏÑ§Ï†ï
fig.suptitle('Ïó¨Îü¨ Í∑∏ÎûòÌîÑ 2Ï∞®ÏõêÏúºÎ°ú ÎÇòÏó¥ÌïòÍ∏∞', fontsize=20, fontweight='bold')

## (3) Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞ Î∞è Íæ∏ÎØ∏Í∏∞
for i in range(2):
    for j in range(3):
        # Îç∞Ïù¥ÌÑ∞
        numbers = i*10 + 3*(j+1)*np.random.standard_normal(10000)  # ÌëúÏ§ÄÏ†ïÍ∑úÎ∂ÑÌè¨Î•º ÌôïÎ•†Î∞ÄÎèÑÌï®ÏàòÎ°ú ÌïòÏó¨ ÎÇúÏàòÎ•º 10000Í∞ú ÏÉùÏÑ±ÌïòÍ≥† Ï¢åÌëúÏóê Îî∞Îùº ÌäπÏ†ïÌïú ÏàòÎ•º Í≥±ÌïòÍ±∞ÎÇò ÎçîÌï®
        # Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
        axes[i, j].hist(numbers, bins=100, density=True, color='lime')  # Í∞íÏùò Î≤îÏúÑÎ•º 100Í∞úÏùò Íµ¨Í∞ÑÏúºÎ°ú ÎÇòÎàÑÏñ¥ ÎÇúÏàòÏùò ÎπàÎèÑÏàòÎ•º ÌûàÏä§ÌÜ†Í∑∏Îû®ÏúºÎ°ú Í∑∏Î¶º
        # Í∑∏ÎûòÌîÑ Íæ∏ÎØ∏Í∏∞
        axes[i, j].set_title("mean:"+str(i*10)+", std:"+str(3*(j+1)), fontsize=20)  # Îç∞Ïù¥ÌÑ∞Ïùò ÌèâÍ∑†Í∞íÍ≥º ÌëúÏ§ÄÌé∏Ï∞®Í∞íÏùÑ Ï†úÎ™©Ïóê Ï∂îÍ∞Ä
        axes[i, j].grid(axis='both', color='purple', linestyle=':', linewidth=1)  # Í≤©Ïûê ÏÑ§Ï†ï
        axes[i, j].set_facecolor('gray')  # axesÏùò Î∞îÌÉïÏÉâ ÏÑ§Ï†ï
        
plt.show()  # figureÎ•º Ï∂úÎ†•
```
