> ‚ë°  `plt.show()` Î•º ÌÜµÌï¥ figureÎ•º Ï∂úÎ†•Ìï† Ïàò ÏûàÏäµÎãàÎã§. 

### **(1) Î¶¨Ïä§Ìä∏ Ïù¥Ïö©ÌïòÍ∏∞**

> List ÌòïÌÉúÏùò Îç∞Ïù¥ÌÑ∞Î•º Îã§ÏùåÍ≥º Í∞ôÏù¥ `plot` Ìï®ÏàòÎ°ú Í∫æÏùÄÏÑ† Í∑∏ÎûòÌîÑÎ°ú ÎÇòÌÉÄÎÇº Ïàò ÏûàÏäµÎãàÎã§.  
>
> `plt.plot(List_X, List_Y)`  
> : List_X (Ïà´ÏûêÎÇò Î¨∏ÏûêÎ•º Îã¥ÏùÄ Î¶¨Ïä§Ìä∏) Ïóê Îî∞Î•∏ List_Y (Ïà´ÏûêÎ•º Îã¥ÏùÄ Î¶¨Ïä§Ìä∏) Îç∞Ïù¥ÌÑ∞Î•º Í∫æÏùÄÏÑ† Í∑∏ÎûòÌîÑÎ°ú Í∑∏Î¶º
>
> Í∑∏ÎûòÌîÑÎ•º Ï∂úÎ†•ÌïòÍ∏∞ ÏúÑÌï¥ÏÑúÎäî Î∞òÎìúÏãú Îã§Ïùå ÏΩîÎìúÎ•º Ïó∞Ïù¥Ïñ¥ Ïã§ÌñâÌï¥ÏïºÌï©ÎãàÎã§.  
> `plt.show()`

```python
# ÏòàÏ†ú 1: List Îç∞Ïù¥ÌÑ∞Î°ú Í∫æÏùÄÏÑ† Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞

X1 = ['1Ïùº', '2Ïùº', '3Ïùº', '4Ïùº', '5Ïùº', '6Ïùº', '7Ïùº']
Y1 = [50, 120, 36, 21, 50, 19, 95]

plt.plot(X1, Y1)  # Îπà figureÏóê Í∫æÏùÄÏÑ† Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶º
plt.show()  # figureÎ•º Ï∂úÎ†•
```

> Í∞Å Îç∞Ïù¥ÌÑ∞Î•º ÎÇòÌÉÄÎÇ¥Îäî Ï¢åÌëúÎ•º ÎèãÎ≥¥Ïù¥Í≤å ÌëúÏãúÌïòÍ≥† Ïã∂Îã§Î©¥ Îã§ÏùåÍ≥º Í∞ôÏù¥ ÏÑ†Ïùò Ïä§ÌÉÄÏùºÏùÑ Î≥ÄÍ≤ΩÌï† Ïàò ÏûàÏäµÎãàÎã§.  
>
> `plt.plot(data, 'o-')`  
>: ÏòÅÏñ¥ ÏïåÌååÎ≤≥ 'o'Îäî Îç∞Ïù¥ÌÑ∞ Ï¢åÌëúÏóê Îë•Í∑ºÏ†êÏùÑ Ï∞çÎäîÎã§Îäî ÏùòÎØ∏Ïù¥Í≥† '-'Îäî Ï†êÎì§ÏùÑ Ïã§ÏÑ†ÏúºÎ°ú ÏûáÎäîÎã§Îäî ÏùòÎØ∏ÏûÖÎãàÎã§.

```python
# ÏòàÏ†ú 2: List Îç∞Ïù¥ÌÑ∞Î°ú Í∫æÏùÄÏÑ† Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í≥† Îç∞Ïù¥ÌÑ∞ Ï¢åÌëúÏóê Ï†ê Ï∞çÍ∏∞

plt.plot(X1, Y1, 'o-')  # Îπà figureÏóê Í∫æÏùÄÏÑ† Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶º
plt.show()  # figureÎ•º Ï∂úÎ†•
```

    üë©‚Äçüè´ <strong>NumPyÎûÄ?</strong>

    - NumPyÏùò array Îç∞Ïù¥ÌÑ∞ Íµ¨Ï°∞Îäî listÏôÄ ÎπÑÏä∑Ìïú ÌòïÌÉúÎ•º ÎùÑÍ≥† ÏûàÏäµÎãàÎã§.</br>
    - import numpy as np Î°ú Ï∂ïÏïΩÌïòÏó¨ ÏÑ†Ïñ∏Ìï¥ ÏÇ¨Ïö©Ìï©ÎãàÎã§.

```python
# ÏòàÏ†ú 3: Numpy.array Îç∞Ïù¥ÌÑ∞Î°ú Í∫æÏùÄÏÑ† Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞

import numpy as np  # numpy ÎùºÏù¥Î∏åÎü¨Î¶¨Î•º ÏÇ¨Ïö©ÌïòÍ∏∞ ÏúÑÌï¥ npÎ°ú Ï∂ïÏïΩÌïòÏó¨ Î∂àÎü¨ÏòµÎãàÎã§.

X2 = np.arange(0, 7, 0.1)  # 0 Î∂ÄÌÑ∞ 0.1 Í∞ÑÍ≤©ÏúºÎ°ú 7 ÎØ∏ÎßåÍπåÏßÄÏùò ÏàòÎ•º arrayÏóê Ï†ÄÏû•
Y2 = np.sin(X2)  # X2 ÏïàÏùò Ïàò xÏóê ÎåÄÌï¥ÏÑú sin(x) Í∞íÏùÑ arrayÏóê Ï†ÄÏû•

print(X2)
print(Y2)

plt.plot(X2, Y2)  # Îπà figureÏóê Í∫æÏùÄÏÑ† Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶º
plt.show()  # figureÎ•º Ï∂úÎ†•
```

> Î¨¥ÏóáÏùÑ ÎÇòÌÉÄÎÇ¥Îäî Í∑∏ÎûòÌîÑÏù∏ÏßÄ ÌëúÌòÑÌïòÍ≥† Ïã∂Îã§Î©¥ Îã§ÏùåÍ≥º Í∞ôÏù¥ Í∑∏ÎûòÌîÑÏùò Ï†úÎ™©ÏùÑ Ï†ïÌï† Ïàò ÏûàÏäµÎãàÎã§.
>
> `plt.title("Graph Title")`
> : figureÏóê Graph Title Ïù¥ÎûÄ Ï†úÎ™©ÏùÑ ÏîÄ

```python
# ÏòàÏ†ú 4: Numpy.array Îç∞Ïù¥ÌÑ∞Î°ú Í∫æÏùÄÏÑ† Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í≥† Ï†úÎ™© Ïì∞Í∏∞

plt.plot(X2, Y2)  # Îπà figureÏóê Í∫æÏùÄÏÑ† Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶º
plt.title("y = sin(x)")  # figureÏùò Ï†úÎ™©ÏùÑ ÏûëÏÑ± 
plt.show()  # figureÎ•º Ï∂úÎ†•
```
